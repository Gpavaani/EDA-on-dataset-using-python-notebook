```python
###FINDING OUTLIERS
df.head()
```

```python
sns.boxplot(df['Price'],orient="v")
```

```python
#finding sale price outles
iqr=np.quantile(df['Price'],0.75)-np.quantile(df['Price'],0.25)
lower_bound=np.quantile(df['Price'],0.25)-(1.5*iqr)
upper_bound=np.quantile(df['Price'],0.75)+(1.5*iqr)
outliers=df[(df['Price']<lower_bound)|(df['Price']>upper_bound)]
outliers
#605 outliers
```

```python
#mean and median
outliers['Price'].mean()
```

```python
outliers['Price'].median()
```

```python
#delete outliers
df=df.drop(outliers.index)
df
```

```python
sns.boxplot(df['Price'])
#compare this with 1st one to see the differnce
```
